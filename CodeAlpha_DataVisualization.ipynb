```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv("fitness_tracker.csv")
```

```python
# Show first few rows
print(data.head())
```

```python
print(data.columns)
```

```python
print(data.dtypes)
```

```python
print(data.shape)   # Example: (100, 6)
```

```python
plt.figure(figsize=(10,5))
plt.plot(data["Date"], data["Steps"], marker="o", color="blue")
plt.title("Steps Over Time")
plt.xlabel("Date")
plt.ylabel("Steps")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(x="Workout_Minutes", y="Calories_Burned", data=data, hue="Workout_Type", palette="Set2")
plt.title("Calories Burned vs Workout Duration")
plt.show()
```

```python
sns.boxplot(x="Mood", y="Sleep_Hours", data=data, palette="coolwarm")
plt.title("Sleep Hours Distribution by Mood")
plt.show()
```

```python
sns.histplot(data["Heart_Rate"], kde=True, color="red")
plt.title("Heart Rate Distribution")
plt.xlabel("Heart Rate (bpm)")
plt.show()
```

```python
avg_protein = data["Protein_Intake_g"].mean()
avg_fat = data["Fat_Intake_g"].mean()
avg_carbs = data["Carbs_Intake_g"].mean()

plt.pie([avg_protein, avg_fat, avg_carbs], 
        labels=["Protein", "Fat", "Carbs"], 
        autopct="%1.1f%%", startangle=90, colors=["#66b3ff","#ff9999","#99ff99"])
plt.title("Average Macronutrient Intake")
plt.show()
```

```python
sns.scatterplot(x="Sleep_Hours", y="Stress_Level", data=data, hue="Mood", palette="viridis")
plt.title("Stress vs Sleep Hours")
plt.show()
```

```python
sns.scatterplot(x="BP_Systolic", y="BP_Diastolic", data=data, hue="Stress_Level", palette="coolwarm")
plt.title("Blood Pressure Levels")
plt.show()
```

```python
plt.plot(data["Date"], data["Oxygen_Level"], marker="o", color="green")
plt.title("Oxygen Level Over Time")
plt.xlabel("Date")
plt.ylabel("Oxygen Level (%)")
plt.xticks(rotation=45)
plt.show()
```

```python
corr = data.corr(numeric_only=True)
plt.figure(figsize=(12,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Fitness Data")
plt.show()
```
